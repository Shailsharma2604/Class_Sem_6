#Exercise: Basic Image Manipulations (Resizing, Cropping, Flipping)

Objective:

Resize an image to a specific dimension.

Crop an image by selecting a region of interest (ROI).

Flip an image both horizontally and vertically.

Libraries:

OpenCV: For image loading, manipulation, and saving.

Matplotlib: For displaying images.

```python
# Import necessary libraries
import cv2  # OpenCV for image processing
import numpy as np  # NumPy for array manipulations
import matplotlib.pyplot as plt  # Matplotlib for displaying images

# Step 1: Load an image using OpenCV
image_path = '123.PNG'  # Replace with the path to your image
image = cv2.imread(image_path)  # Reading the image as a NumPy array

# Check if image is loaded correctly
if image is None:
    print("Error loading image.")
else:
    print(f"Image loaded successfully. Shape: {image.shape}")

# Step 2: Display the original image using Matplotlib
# Convert the color from BGR (OpenCV format) to RGB (Matplotlib format)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Plotting the image
plt.figure(figsize=(8, 6))
plt.imshow(image_rgb)
plt.title("Original Image")
plt.axis('off')  # Hide axes
plt.show()

# Step 3: Resizing the Image
# Resizing the image to a specific width and height (for example, 400x400)
resized_image = cv2.resize(image, (400, 400))  # New size: 400x400

# Display resized image
resized_image_rgb = cv2.cvtColor(resized_image, cv2.COLOR_BGR2RGB)
plt.imshow(resized_image_rgb)
plt.title("Resized Image (400x400)")
plt.axis('off')
plt.show()

# Step 4: Cropping the Image (Region of Interest - ROI)
# Define a region of interest (ROI) by slicing the image array
# For example, crop a region from row 50 to 150, and column 100 to 300
roi = image[50:150, 100:300]

# Display cropped image
roi_rgb = cv2.cvtColor(roi, cv2.COLOR_BGR2RGB)
plt.imshow(roi_rgb)
plt.title("Cropped Image (ROI: 50:150, 100:300)")
plt.axis('off')
plt.show()

# Step 5: Flipping the Image
# Horizontal flip (flip code = 1)
flipped_horizontal = cv2.flip(image, 1)

# Vertical flip (flip code = 0)
flipped_vertical = cv2.flip(image, 0)

# Display flipped images
flipped_horizontal_rgb = cv2.cvtColor(flipped_horizontal, cv2.COLOR_BGR2RGB)
flipped_vertical_rgb = cv2.cvtColor(flipped_vertical, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(12, 6))

# Horizontal flip
plt.subplot(1, 2, 1)
plt.imshow(flipped_horizontal_rgb)
plt.title("Flipped Horizontally")
plt.axis('off')

# Vertical flip
plt.subplot(1, 2, 2)
plt.imshow(flipped_vertical_rgb)
plt.title("Flipped Vertically")
plt.axis('off')

plt.show()

# Step 6: Saving the Modified Images
cv2.imwrite('resized_image.jpg', resized_image)  # Save the resized image
cv2.imwrite('cropped_image.jpg', roi)  # Save the cropped image
cv2.imwrite('flipped_horizontal_image.jpg', flipped_horizontal)  # Save horizontal flipped image
cv2.imwrite('flipped_vertical_image.jpg', flipped_vertical)  # Save vertical flipped image
```

Explanation of the Code:

Loading the Image:

The image is loaded using cv2.imread() and stored as a NumPy array. The shape of the image (height, width, number of channels) is printed to ensure it's loaded correctly.

Displaying the Image:

The image is displayed using Matplotlib, but OpenCV loads images in BGR format (Blue, Green, Red). To display the image correctly in RGB format (Red, Green, Blue), we use cv2.cvtColor() to convert it before plotting.

Resizing the Image:

The cv2.resize() function is used to resize the image to a specific dimension, here chosen as 400x400 pixels. This changes the size of the image while maintaining the pixel data.

Resizing might distort the image if the aspect ratio is not maintained, so make sure the width and height are chosen carefully.

Cropping the Image (Region of Interest):

Image cropping is achieved through slicing the NumPy array. In this case, the region from rows 50 to 150 and columns 100 to 300 is selected. This allows you to focus on specific parts of the image.

The sliced portion (ROI) is then displayed separately.

Flipping the Image:

Horizontal flip is done using cv2.flip(image, 1), where 1 is the flip code for horizontal flipping.

Vertical flip is done using cv2.flip(image, 0), where 0 is the flip code for vertical flipping.

The flipped images are displayed side by side for comparison.

Saving the Modified Images:

The modified images (resized, cropped, and flipped) are saved using cv2.imwrite().

This saves the images to disk with the specified filenames.

Cropped Image: A cropped region (ROI) of the original image.
