Problem Statement:

Objective:

You are given an image that contains a few objects (e.g., shapes or text) on a background. Your task is to perform the following:

Edge Detection:

 Apply both Sobel and Canny edge detectors to identify the boundaries of the objects.

Feature Enhancement:

Enhance the features of the image by applying histogram equalization and contrast adjustment techniques.

Thresholding:

Perform both global thresholding and adaptive thresholding to segment the image into foreground and background.

In this practical, we will:

Load the image.

Apply edge detection using the Sobel and Canny methods.

Enhance the image using histogram equalization and contrast adjustment.

Segment the image using global and adaptive thresholding.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Load the image
image = cv2.imread('123.PNG')  # Replace with a path to your image
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)  # Convert to grayscale for edge detection and thresholding

# Step 2: Apply Sobel Edge Detection
sobel_x = cv2.Sobel(gray_image, cv2.CV_64F, 1, 0, ksize=3)  # Sobel in X direction
sobel_y = cv2.Sobel(gray_image, cv2.CV_64F, 0, 1, ksize=3)  # Sobel in Y direction
sobel_edges = cv2.magnitude(sobel_x, sobel_y)  # Combine the X and Y gradients to get magnitude

# Step 3: Apply Canny Edge Detection
canny_edges = cv2.Canny(gray_image, 100, 200)  # Use low and high threshold for Canny

# Step 4: Feature Enhancement - Histogram Equalization
equalized_image = cv2.equalizeHist(gray_image)  # Perform histogram equalization to enhance contrast

# Step 5: Feature Enhancement - Contrast Adjustment
# Simple contrast adjustment: Multiply pixel values by a constant (e.g., 1.5) to enhance contrast
contrast_adjusted = cv2.convertScaleAbs(gray_image, alpha=1.5, beta=0)

# Step 6: Global Thresholding
_, global_thresholded = cv2.threshold(gray_image, 127, 255, cv2.THRESH_BINARY)

# Step 7: Adaptive Thresholding
adaptive_thresholded = cv2.adaptiveThreshold(gray_image, 255, cv2.ADAPTIVE_THRESH_MEAN_C,
                                            cv2.THRESH_BINARY, 11, 2)

# Step 8: Display the results
plt.figure(figsize=(12, 12))

# Original Image
plt.subplot(3, 3, 1)
plt.imshow(gray_image, cmap='gray')
plt.title("Original Grayscale Image")
plt.axis('off')

# Sobel Edge Detection
plt.subplot(3, 3, 2)
plt.imshow(sobel_edges, cmap='gray')
plt.title("Sobel Edge Detection")
plt.axis('off')

# Canny Edge Detection
plt.subplot(3, 3, 3)
plt.imshow(canny_edges, cmap='gray')
plt.title("Canny Edge Detection")
plt.axis('off')

# Histogram Equalization
plt.subplot(3, 3, 4)
plt.imshow(equalized_image, cmap='gray')
plt.title("Histogram Equalization")
plt.axis('off')

# Contrast Adjustment
plt.subplot(3, 3, 5)
plt.imshow(contrast_adjusted, cmap='gray')
plt.title("Contrast Adjustment")
plt.axis('off')

# Global Thresholding
plt.subplot(3, 3, 6)
plt.imshow(global_thresholded, cmap='gray')
plt.title("Global Thresholding")
plt.axis('off')

# Adaptive Thresholding
plt.subplot(3, 3, 7)
plt.imshow(adaptive_thresholded, cmap='gray')
plt.title("Adaptive Thresholding")
plt.axis('off')

plt.tight_layout()
plt.show()
```

Code Explanation:

Image Loading:

We load the image using cv2.imread() and convert it to grayscale using cv2.cvtColor() since edge detection and thresholding typically work on grayscale images.

Edge Detection:

Sobel Edge Detection:

We apply the Sobel operator using cv2.Sobel() to compute gradients in both the X and Y directions (sobel_x and sobel_y). The magnitude of the gradients is calculated using cv2.magnitude() to obtain the final edge map.

Canny Edge Detection:

We use cv2.Canny() with low and high threshold values (100 and 200 in this case) to detect edges based on intensity gradients.

Feature Enhancement:

Histogram Equalization:

 We apply cv2.equalizeHist() to enhance the contrast of the grayscale image, spreading pixel intensities more evenly across the available range.

Contrast Adjustment:

 We apply simple contrast adjustment using cv2.convertScaleAbs(), where alpha=1.5 scales the pixel intensities, increasing contrast.

Thresholding:

Global Thresholding:

 We use cv2.threshold() to apply a fixed threshold (127) across the entire image. Any pixel with a value greater than or equal to 127 is set to 255 (white), and the rest are set to 0 (black).

Adaptive Thresholding:

 We apply cv2.adaptiveThreshold() to calculate a local threshold for each pixel based on its neighborhood (with a block size of 11 and a constant of 2).

Displaying the Results:

We use matplotlib.pyplot to display the results in a 3x3 grid.

Each subplot shows the output of a different technique: Sobel and Canny edge detection, histogram equalization, contrast adjustment, global thresholding, and adaptive thresholding.

Expected Output:

Original Grayscale Image:

The grayscale version of the input image.
Sobel Edge Detection: Detected edges using the Sobel operator, highlighting regions with high gradient changes.

Canny Edge Detection:

Fine edges detected with the Canny edge detector, which typically provides cleaner and more precise edges.

Histogram Equalization:

A high-contrast version of the image, where dark and bright regions are enhanced.

Contrast Adjustment:

 A contrast-enhanced image that makes the features more distinct.
Global Thresholding: A binary image where regions with pixel values above the threshold are white, and others are black.

Adaptive Thresholding:

 A locally thresholded image where different regions may have different binary thresholds.

Conclusion:

Edge Detection:

 Sobel and Canny are both effective edge detectors, with Canny generally providing more refined edges.

Feature Enhancement:

 Histogram equalization and contrast adjustment help to highlight features and improve image quality.

Thresholding:

Global thresholding works well in uniform lighting conditions, while adaptive thresholding is more flexible and handles varying illumination effectively.
