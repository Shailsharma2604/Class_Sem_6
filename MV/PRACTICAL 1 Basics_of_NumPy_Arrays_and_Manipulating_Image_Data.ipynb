#Practical for Basics of NumPy Arrays and Manipulating Image Data

Objective:

Understand how images are represented as NumPy arrays.

Access and manipulate pixel values.

Perform basic image processing tasks by directly modifying the NumPy array representing the image.

Libraries:

OpenCV: For loading and saving images.

NumPy: For array manipulation and performing operations on image data.

Matplotlib: For visualizing images.

```python
# Import necessary libraries
import cv2  # OpenCV for image processing
import numpy as np  # NumPy for array manipulations
import matplotlib.pyplot as plt  # Matplotlib for displaying images

# Step 1: Load an image using OpenCV
image_path = '123.PNG'  # Replace with the path to your image
image = cv2.imread(image_path)  # Reading the image as a NumPy array

# Check if image is loaded correctly
if image is None:
    print("Error loading image.")
else:
    print(f"Image loaded successfully. Shape: {image.shape}")

# Step 2: Display the original image using Matplotlib
# Convert the color from BGR (OpenCV format) to RGB (Matplotlib format)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Plotting the image
plt.figure(figsize=(8, 6))
plt.imshow(image_rgb)
plt.title("Original Image")
plt.axis('off')  # Hide axes
plt.show()

# Step 3: Accessing Pixel Values
# Accessing the pixel value at row 100, column 50 (BGR values)
pixel_value = image[100, 50]
print(f"Pixel value at (100, 50): {pixel_value}")

# Accessing individual color channels
blue_channel = image[100, 50, 0]  # Blue channel value
green_channel = image[100, 50, 1]  # Green channel value
red_channel = image[100, 50, 2]  # Red channel value

print(f"Blue channel: {blue_channel}, Green channel: {green_channel}, Red channel: {red_channel}")

# Step 4: Modifying Pixel Values
# Change the pixel at (100, 50) to a specific color (e.g., yellow)
image[100, 50] = [0, 255, 255]  # Set pixel to Yellow (BGR)

# Plotting the modified image
image_rgb_modified = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb_modified)
plt.title("Modified Image (Pixel Changed to Yellow)")
plt.axis('off')
plt.show()

# Step 5: Slicing and Region of Interest (ROI)
# Extracting a part of the image (Region of Interest)
roi = image[50:150, 100:200]  # Crop a section from row 50 to 150 and column 100 to 200

# Display the extracted ROI
roi_rgb = cv2.cvtColor(roi, cv2.COLOR_BGR2RGB)
plt.imshow(roi_rgb)
plt.title("Region of Interest (ROI)")
plt.axis('off')
plt.show()

# Step 6: Changing a Region of Interest
# Let's change the ROI to a solid color (e.g., Blue)
image[50:150, 100:200] = [255, 0, 0]  # Change to Blue (BGR)

# Display the image with modified ROI
image_rgb_modified_roi = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb_modified_roi)
plt.title("Image with Modified ROI (Blue)")
plt.axis('off')
plt.show()

# Step 7: Image Arithmetic - Brightness Adjustment
# Increase the brightness by adding a constant to all pixel values
brightened_image = cv2.convertScaleAbs(image, alpha=1, beta=50)  # alpha is the scaling factor, beta is the brightness offset

# Display the brightened image
brightened_image_rgb = cv2.cvtColor(brightened_image, cv2.COLOR_BGR2RGB)
plt.imshow(brightened_image_rgb)
plt.title("Brightened Image")
plt.axis('off')
plt.show()

# Step 8: Applying a Simple Filter (Smoothing with a Kernel)
# Create a simple averaging kernel for smoothing
kernel = np.ones((5, 5), np.float32) / 25  # 5x5 averaging kernel

# Apply filter using cv2.filter2D()
smoothed_image = cv2.filter2D(image, -1, kernel)

# Display the smoothed image
smoothed_image_rgb = cv2.cvtColor(smoothed_image, cv2.COLOR_BGR2RGB)
plt.imshow(smoothed_image_rgb)
plt.title("Smoothed Image (Averaging Filter)")
plt.axis('off')
plt.show()

# Step 9: Saving the Modified Image
cv2.imwrite('modified_image.jpg', image)  # Save the image after modification (e.g., pixel change and ROI)
```

#Explanation of the Code:

Loading and Displaying the Image:

The image is loaded into a NumPy array using cv2.imread(). The shape of the image is printed to confirm it has loaded correctly (height, width, and number of channels).

We convert the image from BGR (OpenCVs default format) to RGB (required by Matplotlib) using cv2.cvtColor() for display purposes.

Accessing Pixel Values:

You can access the pixel values at any specific location using NumPy array indexing. In the example, we access the pixel at row 100 and column 50 and print its BGR values.

Each pixel has three values: Blue (channel 0), Green (channel 1), and Red (channel 2), corresponding to the BGR format.

Modifying Pixel Values:

We modify the pixel at location (100, 50) by directly assigning a new BGR value [0, 255, 255], which corresponds to Yellow color.

After modification, the image is displayed again using Matplotlib.

Slicing and Region of Interest (ROI):

Image regions (or parts of the image) can be extracted using slicing. In this case, we extract a region from rows 50 to 150 and columns 100 to 200.

The extracted region (ROI) is displayed separately.

Changing a Region of Interest:

You can modify any part of the image (ROI) by directly assigning new pixel values to the sliced region. We change the ROI to Blue color ([255, 0, 0] in BGR format) and display the modified image.

Image Arithmetic - Brightness Adjustment:

To adjust the brightness, we use cv2.convertScaleAbs(), which scales the pixel values and adds an offset (beta). By adding beta=50, we increase the brightness of the entire image.

Applying a Simple Filter (Smoothing):

We apply an averaging filter (a 5x5 kernel) to the image to perform smoothing. The kernel is a matrix filled with values 1/25, and cv2.filter2D() applies this kernel to the image.

This operation blurs the image, which can be used to reduce noise.
Saving the Image:

After all the manipulations, the modified image can be saved to a file using cv2.imwrite().
