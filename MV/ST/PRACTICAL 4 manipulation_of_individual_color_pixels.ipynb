Key Manipulations:

Blue Channel Manipulation:

Added 50 to all pixel values in the Blue channel.

Used np.clip to ensure that pixel values remain in the range [0, 255].


Green Channel Manipulation:

Set the top-left quarter of the Green channel to 0 (black), creating a noticeable change in the region.

Red Channel Manipulation:

Inverted the pixel intensity values of the Red channel using the formula 255 - pixel_value.

```python
# Import necessary libraries
import cv2  # OpenCV library for image processing
import numpy as np  # NumPy library for numerical operations
import matplotlib.pyplot as plt  # Matplotlib library for visualization

# Load the image from a specified file path
image_path = "123.PNG"  # Replace with the path to your image file
image = cv2.imread(image_path)  # Read the image in BGR format (default format in OpenCV)

# Check if the image was loaded successfully
if image is None:  # If the image is not loaded (e.g., incorrect path or missing file)
    print("Error: Unable to load the image. Please check the file path.")
    exit()  # Exit the program if the image cannot be loaded

# Split the image into its Blue, Green, and Red channels
blue_channel, green_channel, red_channel = cv2.split(image)

# Manipulate the Blue channel
# Example: Increase the intensity of all blue pixels by 50 (cap at 255)
blue_channel = np.clip(blue_channel + 50, 0, 255)

# Manipulate the Green channel
# Example: Set all pixels in the top-left quarter of the image to 0
height, width = green_channel.shape
green_channel[:height // 2, :width // 2] = 0

# Manipulate the Red channel
# Example: Invert the red channel intensity
red_channel = 255 - red_channel

# Merge the manipulated channels back into a single image
manipulated_image = cv2.merge((blue_channel, green_channel, red_channel))

# Create a figure for visualizing the results
plt.figure(figsize=(15, 8))  # Set the figure size for better visibility

# Display the original image
plt.subplot(1, 2, 1)  # Create a 1x2 grid and select the 1st position for the original image
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))  # Convert BGR to RGB for correct color display
plt.title("Original Image")  # Add a title to the plot
plt.axis("off")  # Remove axes for a cleaner display

# Display the manipulated image
plt.subplot(1, 2, 2)  # Select the 2nd position for the manipulated image
plt.imshow(cv2.cvtColor(manipulated_image, cv2.COLOR_BGR2RGB))  # Convert BGR to RGB for correct color display
plt.title("Manipulated Image")  # Add a title to indicate the manipulated image
plt.axis("off")  # Remove axes for a cleaner display

# Adjust layout to ensure titles and plots don't overlap
plt.tight_layout()

# Show the final plot with both images
plt.show()
```
