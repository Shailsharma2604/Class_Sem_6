**Splitting Channels >>**

cv2.split() is used to split coloured/multi-channel image into separate single-channel images.

The cv2.split() is an expensive operation in terms of performance(time). The order of the output vector of arrays depends on the order of channels of the input image.


Syntax: cv2.split(m[, mv])


Parameters:


m: Input multi-channel array
mv: Output vector of arrays



```python
# Import necessary libraries
import cv2  # OpenCV library for image processing
import matplotlib.pyplot as plt  # Matplotlib library for visualization

# Load the image from a specified file path
image_path = "123.PNG"  # Replace with the path to your image file
image = cv2.imread(image_path)  # Read the image in BGR format (default format in OpenCV)

# Check if the image was loaded successfully
if image is None:  # If the image is not loaded (e.g., incorrect path or missing file)
    print("Error: Unable to load the image. Please check the file path.")
    exit()  # Exit the program if the image cannot be loaded

# Split the image into its Blue, Green, and Red channels
# Each channel represents the intensity values for that specific color
blue_channel, green_channel, red_channel = cv2.split(image)

# Create a figure for visualizing the results
plt.figure(figsize=(12, 8))  # Set the figure size for better visibility

# Display the original image
plt.subplot(2, 2, 1)  # Create a 2x2 grid and select the 1st position for the original image
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))  # Convert BGR to RGB for correct color display
plt.title("Original Image")  # Add a title to the plot
plt.axis("off")  # Remove axes for a cleaner display

# Display the Blue channel
plt.subplot(2, 2, 2)  # Select the 2nd position in the grid for the Blue channel
plt.imshow(blue_channel, cmap="gray")  # Display the Blue channel in grayscale
plt.title("Blue Channel")  # Add a title to indicate this is the Blue channel
plt.axis("off")  # Remove axes for a cleaner display

# Display the Green channel
plt.subplot(2, 2, 3)  # Select the 3rd position in the grid for the Green channel
plt.imshow(green_channel, cmap="gray")  # Display the Green channel in grayscale
plt.title("Green Channel")  # Add a title to indicate this is the Green channel
plt.axis("off")  # Remove axes for a cleaner display

# Display the Red channel
plt.subplot(2, 2, 4)  # Select the 4th position in the grid for the Red channel
plt.imshow(red_channel, cmap="gray")  # Display the Red channel in grayscale
plt.title("Red Channel")  # Add a title to indicate this is the Red channel
plt.axis("off")  # Remove axes for a cleaner display

# Adjust layout to ensure titles and plots don't overlap
plt.tight_layout()

# Show the final plot with all images
plt.show()
```

**Merging Channels >>**

cv2.merge() is used to merge several single-channel images into a colored/multi-channel image.

Syntax: cv2.merge(mv[, dst])


Parameters:


mv: Input vector of matrices to be merged. All matrices must have same size.
dst: Output multi-channel array of size mv[0]. Number of channel will be equal to total no. of channel in matrix array.

```python
# Import necessary libraries
import cv2  # OpenCV library for image processing
import matplotlib.pyplot as plt  # Matplotlib library for visualization

# Load the image from a specified file path
image_path = "123.PNG"  # Replace with the path to your image file
image = cv2.imread(image_path)  # Read the image in BGR format (default format in OpenCV)

# Check if the image was loaded successfully
if image is None:  # If the image is not loaded (e.g., incorrect path or missing file)
    print("Error: Unable to load the image. Please check the file path.")
    exit()  # Exit the program if the image cannot be loaded

# Split the image into its Blue, Green, and Red channels
# Each channel represents the intensity values for that specific color
blue_channel, green_channel, red_channel = cv2.split(image)

# Merge the channels back into a single image
# The order of channels is important: (Blue, Green, Red) for OpenCV
merged_image = cv2.merge((blue_channel, green_channel, red_channel))

# Create a figure for visualizing the results
plt.figure(figsize=(15, 5))  # Set the figure size for better visibility

# Display the original image
plt.subplot(1, 3, 1)  # Create a 1x3 grid and select the 1st position for the original image
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))  # Convert BGR to RGB for correct color display
plt.title("Original Image")  # Add a title to the plot
plt.axis("off")  # Remove axes for a cleaner display

# Display the individual channels (for reference)
plt.subplot(1, 3, 2)  # Select the 2nd position for the split channels visualization
plt.imshow(cv2.merge((blue_channel, green_channel, red_channel)), cmap="gray")
plt.title("Blue ")

# Display the merged image (after combining the channels)
plt.subplot(1, 3, 3)  # Select the 3rd position for the merged image
plt.imshow(cv2.cvtColor(merged_image, cv2.COLOR_BGR2RGB))  # Convert BGR to RGB for display
plt.title("Merged Image")  # Add a title to indicate the merged image
plt.axis("off")  # Remove axes for a cleaner display

# Adjust layout to ensure titles and plots don't overlap
plt.tight_layout()

# Show the final plot with all images
plt.show()
```
