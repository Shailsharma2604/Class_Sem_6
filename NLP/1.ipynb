```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Read the image
image = cv2.imread('13.png')

if image is None:
    print("Could not open or read the image")
else:
    # Resize the image
    image = cv2.resize(image, (800, 500))
    
    # Convert BGR to RGB (OpenCV uses BGR by default, but Matplotlib expects RGB)
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    
    # Display the image using matplotlib
    plt.imshow(image_rgb)
    plt.axis('off')  # Hide axes for a cleaner display
    plt.show()

# Print image dimensions
print(image.shape)
```

```python
# Accessing pixel values
pixel_value = image[100, 150]  # Accessing the pixel at row 100, column 150
print("Pixel value at (100, 150):", pixel_value)

# Accessing individual color channels (BGR)
blue = image[100, 150, 0]
green = image[100, 150, 1]
red = image[100, 150, 2]
print("B:", blue, "G:", green, "R:", red)

# Modifying a pixel
image[100, 150] = [255, 0, 0]  # Setting the pixel to blue
  
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB)) # Convert BGR to RGB
plt.axis('off') # Hide axes
plt.show()
```

```python
image = cv2.resize(image, (800, 500))
# Modify pixel values (example: increase brightness)
image = cv2.convertScaleAbs(image, alpha=1.5, beta=30) # alpha controls contrast, beta controls brightness
cv2.imshow('Image', image)
print(image.shape)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB)) # Convert BGR to RGB
plt.axis('off') # Hide axes
plt.show()
```

```python
# Define the region of interest (ROI)
x_start = 100  # Starting x-coordinate
y_start = 50   # Starting y-coordinate
width = 200    # Width of the ROI
height = 150   # Height of the ROI

roi = image[y_start:y_start+height, x_start:x_start+width]

cv2.imshow('Roi',roi)
plt.imshow(cv2.cvtColor(roi, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

```python
image[y_start:y_start+height, x_start:x_start+width] = [0, 255, 0] #change ROI to green

cv2.imshow('Image',image)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB)) # Convert BGR to RGB
plt.axis('off') # Hide axes
plt.show()
```
